```python
from datasets import load_dataset

data_files = {"train": "SQuAD_it-train.json", "test": "SQuAD_it-test.json"}
squad_it_dataset = load_dataset("json", data_files=data_files, field="data")
squad_it_dataset
```

## Automatic Decompression
There was no need to decompress the files, since hugging face datasets can handle that automatically.

```python
data_files = {"train": "SQuAD_it-train.json.gz", "test": "SQuAD_it-test.json.gz"}
auto_squad_it_dataset = load_dataset("json", data_files=data_files, field="data")
auto_squad_it_dataset
```

## Automatic Download
To go a further step, datasets library can automatically access remote data when provided with its url.

```python
url = "https://github.com/crux82/squad-it/raw/master/"
data_files = {
    "train": url + "SQuAD_it-train.json.gz",
    "test": url + "SQuAD_it-test.json.gz",
}
remote_squad_it_dataset = load_dataset("json", data_files=data_files, field="data")
remote_squad_it_dataset
```
